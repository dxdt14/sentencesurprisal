```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import time
```

```python
#read the text file
file = open("yelp/train.txt", "r", encoding = "utf8")
file_data = file.read()
lines = file_data.splitlines()
file.close()

data = []

#only use senteces that contain the keyword
keyword = "pizza"
for s in lines:
    if keyword in s:
        data.append(s)

#appending a random weird sentence to see if it pops to the top of the sentence rankings        
data.append("Colorless green ideas sleep furiously .")
data.append("Les idées vertes incolores dorment furieusement .")
data.append("Ideas verdes incoloras duermen furiosamente .")
#data.append("Have a nice day !")
```

```python
start = time.time()
IDF = TfidfVectorizer(lowercase=False)
result = IDF.fit_transform(data)
end = time.time()
print(end-start)
```

```python
idf_results = {}
for i, j in zip(IDF.get_feature_names_out(), IDF.idf_):
    idf_results[i] = j
```

```python
# footballers_goals = {'Eusebio': 120, 'Cruyff': 104, 'Pele': 150, 'Ronaldo': 132, 'Messi': 125}

# sorted_footballers_by_goals = sorted(footballers_goals.items(), key=lambda x:x[1])
# print(sorted_footballers_by_goals)
```

```python
sen_surprisals = []
temp = []
#print(data)
for i in data:
    for j in i.split():
        if not j in idf_results.keys():
            continue
        temp.append(idf_results[j])
    sen_surprisals.append(np.mean(temp))
    temp.clear()
print(sen_surprisals)
```

```python
dict_sen_surprisals = {}
for i in range(len(sen_surprisals)):
    dict_sen_surprisals[data[i]] = sen_surprisals[i]
```

```python
sorted_idf_results = sorted(dict_sen_surprisals.items(), key= lambda x:x[1], reverse=True)
sorted_sentences = []
for i in sorted_idf_results:
    sorted_sentences.append(i[0])
```

```python
print(sorted_sentences)
```

```python
with open("IDF Sentence_Rankings", 'w') as f:
    for i in sorted_sentences:
        f.write(i + "\n")
    f.close()
```
